## Dataset Description
This is the dataset known as Online Shopping Dataset, which contains information about the transactions and interactions that occur with a customer in an online shopping setting. It covers customer demographics, the description of the products, characteristics of purchases, and engagement characteristics with promotions among others:

1. **Customer Information**:  
   - **CustomerID**: Unique identifier for each customer.  
   - **Gender**: Gender of the customer.  
   - **Location**: Geographic location of the customer.  
   - **Tenure_Months**: Duration (in months) of the customer's association with the platform.  

2. **Transaction Details**:  
   - **Transaction_ID**: Unique identifier for each transaction.  
   - **Transaction_Date**: Date when the transaction occurred.  
   - **Avg_Price**: Average price of the purchased product(s).  
   - **Delivery_Charges**: Shipping charges for the order.  
   - **GST**: Goods and Services Tax applied to the transaction.  

3. **Product Information**:  
   - **Product_SKU**: Stock Keeping Unit, a unique identifier for each product.  
   - **Product_Description**: A detailed description of the product.  
   - **Product_Category**: The category to which the product belongs.  

4. **Promotional Engagement**:  
   - **Coupon_Status**: Indicates whether a coupon was used, not used, or clicked during the transaction.  
   - **Coupon_Code**: Code of the coupon applied to the transaction.  
   - **Discount_pct**: Percentage of the discount applied.  

5. **Spending Insights**:  
   - **Offline_Spend**: Total amount spent by the customer in offline shopping.  
   - **Online_Spend**: Total amount spent by the customer in online shopping.  

6. **Time-Based Features**:  
   - **Date**: Date in alternative formats for flexibility in time analysis.  
   - **Month**: Numeric representation of the month for seasonal trend analysis.  

This dataset records individual transactions and their associated details, allowing for in-depth analysis of customer behavior, product trends, and promotional effectiveness. Its structure supports evaluating time-based trends, spending patterns, and customer segmentation, providing valuable insights into online shopping dynamics.

### Handling Outliers Using IQR
Outliers can skew the results of our analysis. We will handle them using the Interquartile Range (IQR) method. 
Any data points outside the range of [Q1 - 1.5*IQR, Q3 + 1.5*IQR] will be considered outliers and removed.


```python

# Set the size of the plot
plt.figure(figsize=(14, 6))

# Box plot for 'Avg_Price' before removing outliers
plt.subplot(1, 2, 1)
sns.boxplot(x=df['Avg_Price'], color='lightblue')
plt.xlabel('Average Price')
plt.title('Box Plot of Average Price (Before Handling Outliers)')

# Box plot for 'Quantity' before removing outliers
plt.subplot(1, 2, 2)
sns.boxplot(x=df['Quantity'], color='lightcoral')
plt.xlabel('Quantity')
plt.title('Box Plot of Quantity (Before Handling Outliers)')

# Display the plot
plt.tight_layout()
plt.show()

```

```python
# Function to handle outliers using IQR with a larger multiplier
def remove_outliers_iqr(df, column, multiplier=1.5):
    Q1, Q3 = df[column].quantile([0.25, 0.75])
    IQR = Q3 - Q1
    lower_bound = Q1 - multiplier * IQR
    upper_bound = Q3 + multiplier * IQR
    print(f"Q1: {Q1}, Q3: {Q3}, IQR: {IQR}, Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Remove outliers in 'Quantity' and 'Avg_Price' using a larger IQR multiplier
df = remove_outliers_iqr(df, 'Quantity', multiplier=2.5)
df = remove_outliers_iqr(df, 'Avg_Price', multiplier=2.5)

# Check the range of 'Quantity' before and after handling outliers
print(f"Quantity range before: {df['Quantity'].min()} to {df['Quantity'].max()}")
print(f"Quantity range after: {df['Quantity'].min()} to {df['Quantity'].max()}")
```

### Customer Segmentation

#### RFM analysis (Recency, Frequency, and Monetary value). This involves:

##### Recency: How recently a customer made a purchase.
##### Frequency: How often they purchase.
##### Monetary Value: How much they spen

```python
# Recency: Calculate days since last purchase
df['Transaction_Date'] = pd.to_datetime(df['Transaction_Date'], errors='coerce')
latest_date = df['Transaction_Date'].max()
df['Recency'] = (latest_date - df['Transaction_Date']).dt.days

# Frequency: Count number of transactions per customer
df['Frequency'] = df.groupby('CustomerID')['Transaction_ID'].transform('count')

# Monetary Value: Total spend per customer
df['Monetary'] = df.groupby('CustomerID')['Total_Sales'].transform('sum')

# Plotting Recency, Frequency, and Monetary Value
fig, ax = plt.subplots(1, 3, figsize=(18, 6))

# Recency plot
sns.histplot(df['Recency'], bins=30, ax=ax[0], color='blue', kde=True)
ax[0].set_title('Recency Distribution')
ax[0].set_xlabel('Days Since Last Purchase')

# Frequency plot
sns.histplot(df['Frequency'], bins=30, ax=ax[1], color='green', kde=True)
ax[1].set_title('Frequency Distribution')
ax[1].set_xlabel('Number of Transactions')

# Monetary Value plot
sns.histplot(df['Monetary'], bins=30, ax=ax[2], color='purple', kde=True)
ax[2].set_title('Monetary Value Distribution')
ax[2].set_xlabel('Total Spend')

plt.tight_layout()
plt.show()
```

## RFM Features & Customer Lifetime Value (CLV)

### RFM Segmentation

#### It creates RFM segments by assigning quartile ranks to each customer and combining them into an overall segment. The RFM score is the sum of these ranks, providing a summary of customer value.

```python
# Define quantiles for Recency, Frequency, and Monetary
df['R_Quartile'] = pd.qcut(df['Recency'], 4, labels=[4, 3, 2, 1])
df['F_Quartile'] = pd.qcut(df['Frequency'], 4, labels=[1, 2, 3, 4])
df['M_Quartile'] = pd.qcut(df['Monetary'], 4, labels=[1, 2, 3, 4])

# Combine to create RFM Segment
df['RFM_Segment'] = df['R_Quartile'].astype(str) + df['F_Quartile'].astype(str) + df['M_Quartile'].astype(str)

# Calculate RFM Score
df['RFM_Score'] = df[['R_Quartile', 'F_Quartile', 'M_Quartile']].sum(axis=1)

# Display the RFM segmentation results
rfm_summary = df[['CustomerID', 'Recency', 'Frequency', 'Monetary', 'RFM_Segment', 'RFM_Score']].drop_duplicates()
rfm_summary.head()
```

### Customer Lifetime Value (CLV)

#### It will estimate the CLV for each customer based on their average order value and purchase frequency.

```python
# Calculate average order value
df['Avg_Order_Value'] = df['Monetary'] / df['Frequency']

# Calculate purchase frequency
purchase_frequency = df['Frequency'].mean()

# Calculate customer value (CLV) as Avg Order Value * Purchase Frequency
df['CLV'] = df['Avg_Order_Value'] * purchase_frequency

# Display CLV results
clv_summary = df[['CustomerID', 'Avg_Order_Value', 'Frequency', 'CLV']].drop_duplicates()
clv_summary.head()
```

```python
# Convert 'Transaction_Date' to datetime format
df['Transaction_Date'] = pd.to_datetime(df['Transaction_Date'])

# Group data by month and year, summing total sales
df_monthly_sales = df.groupby(df['Transaction_Date'].dt.to_period('M')).agg({'Total_Sales': 'sum'}).reset_index()

# Set the 'Transaction_Date' as the index for time series analysis
df_monthly_sales['Transaction_Date'] = df_monthly_sales['Transaction_Date'].dt.to_timestamp()
df_monthly_sales.set_index('Transaction_Date', inplace=True)

# Preview the monthly sales data
df_monthly_sales.head()
```

```python
# Calculate monthly sales growth
df_monthly_sales['Sales_Growth'] = df_monthly_sales['Total_Sales'].pct_change() * 100

# Preview the sales growth
df_monthly_sales[['Total_Sales', 'Sales_Growth']].head()
```

```python
# Assuming df_monthly_sales has the 'Total_Sales' column with monthly data
# Calculate rolling average with a window size of 3 months (quarterly trend)
df_monthly_sales['Rolling_Avg'] = df_monthly_sales['Total_Sales'].rolling(window=3).mean()

# Plot the original sales and rolling average for trend visualization
plt.figure(figsize=(10, 6))
plt.plot(df_monthly_sales['Total_Sales'], label='Monthly Sales', color='blue')
plt.plot(df_monthly_sales['Rolling_Avg'], label='Rolling Average (3 months)', color='orange', linestyle='--')
plt.title('Monthly Sales and Rolling Average(Line Plot)')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.show()
```
